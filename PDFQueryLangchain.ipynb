```python
from PyPDF2 import PdfReader
from langchain.embeddings.openai import OpenAIEmbeddings 
# embedding vectors
# openAI embeddings measure the relatedness of text strings and used for search, cluster, recommendations, anomaly detection, diversity measurement and classification
from langchain.text_splitter import CharacterTextSplitter 
# content in pdf is split by considering special characters 
# used bcos embeddings have fixed sized of tokens 
from langchain.vectorstores import FAISS
# vector database used with embeddings 
# text data embeddings is stored here 
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
```

```python
import os
os.environ["OPENAI_API_KEY"] = ""
```

```python
reader=PdfReader('LiteratureReview-FinalDraft.pdf')
```

```python
from typing_extensions import Concatenate # to read the lines 
# read text from pdf
raw_text = ''
for i, page in enumerate(reader.pages):
    content = page.extract_text()
    if content:
        raw_text += content
```

```python
print(raw_text)
```

```python
# We need to split the text using Character Text Split such that it should not increse token size
text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 800, # you can play with this 
    chunk_overlap  = 200,
    length_function = len,
)
texts = text_splitter.split_text(raw_text)

print(len(texts)) # total number of texts you are able to get
```

```python
# Download embeddings from OpenAI
embeddings = OpenAIEmbeddings() 
```

```python
document_search = FAISS.from_texts(texts, embeddings) # getting the vectors and storing them
print(document_search)
```

```python
chain = load_qa_chain(OpenAI(), chain_type="stuff") # ask questions
```

```python
# Example 1

query = "What is this paper about?"
docs = document_search.similarity_search(query) # entire embedding in doc search
print(chain.run(input_documents=docs, question=query))
```

```python
# Example 2

query = "What was the limitations addressed in this paper?"
docs = document_search.similarity_search(query) # entire embedding in doc search
print(chain.run(input_documents=docs, question=query))
```
